```python
import pandas as pd
df = pd.read_csv('HW1_data.csv')
dt = pd.DataFrame(df)
print(dt.iat[0,0])
print(dt)
name = []
col = []
for i in range(276):
    s1 = dt.iat[i,0]
    name.append(s1)
names = set(name)
name = list(names)
```

```python
data = pd.DataFrame(columns = ['Name','Sex','Age','HR','Height','Weight','BP'],index = range(len(name)))
for i in range(len(name)):
    data.iat[i,0] = name[i]
for i in range(276):
    d = data[data['Name'] == dt.iat[i,0]].index[0]
    s = dt.iat[i,1]
    data.at[d,s] = dt.iat[i,2]
data
for i in range(50):
    for j in range(2,7):
        if(data.iat[i,j] == data.iat[i,j]):
            data.iat[i,j] = int(data.iat[i,j])
```

```python
avg = pd.DataFrame({'Name':'average'},index = [50])
if((50 in data.index)==False):
    data = pd.concat([data,avg])
for col_name in data.columns:
    if((col_name != 'Name')and (col_name != 'Sex')):
        data.at[50,col_name] = data[col_name].mean(skipna=True)
    #q = data['Age'].idmax(skipna=True)
    #print(data.iat[q,0])
        a = col_name
        b = data[col_name].max(skipna=True)
        print(a,end='')
        print('最大值為:',end=' ')
        print(b)
data
```

```python
cor = []
for i in data['Sex']:
    if(i=='F'):
        cor.append('b')
    else:
        cor.append('r')
import matplotlib.pyplot as plt
plt.scatter(x=data['Height'],y=data['Weight'],c=cor)
#plt.title('身高體重表')
plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
age = [0]*10
for i in data['Age']:
    sel = int(i//10)
    age[sel] += 1

xl = [0]*10
for i in range(10):
    xl[i] = '{i}-{j}'\
           .format(i=(i+1)*10-9,j=(i+1)*10)
plt.bar(xl,age,color='g')
plt.xlabel('age interval')
plt.ylabel('population')
plt.show()
print(age)
```

```python
for i in range(len(age)):
    if(age[i]==0):
        del age[i]
        del xl[i]
plt.pie(age,labels=xl)
plt.legend(loc='right')
plt.show()
```
